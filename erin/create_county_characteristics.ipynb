Create a csv file with the county FIPS code and the following county-level census data:
- 2013 and 2023 RUCC
- 2013 and 2024 Urban Influence Code
- Estimated population in poverty 2023
- Estimated percent of population in poverty 2023
- Estimated median income 2023
- Metro/nonmetro delimination 2023
- Unemployment rate in 2023
- Size of civilian labor force 2023',
- Number Employed 2023
- Number Unemployed 2023
- Census population numbers in 2020
- Economic county topology coddes in 2015

All of this data comes from the USDA Economic research service https://www.ers.usda.gov/data-products/county-level-data-sets

```python
import pandas as pd

poverty_df = pd.read_excel('data/Poverty2023.xlsx', dtype={'FIPS_Code': str}, header=4).rename(
    columns = {
        'FIPS_Code': 'FIPS',
    }
)

poverty_columns = ['FIPS', 'Rural_Urban_Continuum_Code_2013',
                   'Urban_Influence_Code_2013',
                   'Rural_Urban_Continuum_Code_2023',
                   'Urban_Influence_Code_2024',
                   'POVALL_2023',
                   'PCTPOVALL_2023',
                   'MEDHHINC_2023',
]

poverty_df_reduc = poverty_df[poverty_columns]
```

```python
unemployment_df = pd.read_excel('data/Unemployment2023.xlsx', dtype={'FIPS_Code': str}, header=4).rename(
    columns = {
        'FIPS_Code': 'FIPS',
    }
)

unemployment_columns = ['FIPS',
    'Metro_2023',
    'Unemployment_rate_2023',
    'Civilian_labor_force_2023',
    'Employed_2023',
    'Unemployed_2023'
]

unemployment_df_reduc = unemployment_df[unemployment_columns]
```

```python
population_df = pd.read_excel('data/PopulationEstimates.xlsx', dtype={'FIPStxt': str}, header=4).rename(
    columns = {
        'FIPStxt': 'FIPS',
    }
)

population_columns = ['FIPS', 'CENSUS_2020_POP', 'Economic_typology_2015']

population_df_reduc = population_df[population_columns]
```

```python
education_df = pd.read_excel('data/Education2023.xlsx', dtype={'FIPS_Code': str}, header=3).rename(
    columns = {
        'FIPS Code': 'FIPS',
    }
)

education_df
```

```python
all_data = pd.merge(population_df_reduc, poverty_df_reduc, on='FIPS')
all_data = pd.merge(all_data, unemployment_df_reduc, on='FIPS')

all_data.head()
```

```python
all_data.to_csv("../data/county_characteristics.csv", index=False)
```

```python
site_data_dir = Path("../data/IWQIS/site_data/")
site_data_ext = "_all_data.csv"
uid = "WQS0015"

def get_all_data_from_site(site_uid: str):
    return pd.read_csv(Path(site_data_dir, site_uid + site_data_ext))

site_df = get_all_data_from_site(uid)
site_df.head()
    
```

```python
def aggregate_by_interval(df, timestamp_col, value_col, interval, agg_func='mean'):
    """
    Aggregate a time series by a specified interval.
    
    Parameters
    ----------
    df : pd.DataFrame
    timestamp_col : str
        Column containing timestamps in any pandas-parseable format.
    value_col : str or list
        Column(s) to aggregate.
    interval : str
        Pandas offset alias, e.g. '1h', '2h', '1D', '15min', '1W'.
    agg_func : str or callable
        Aggregation function, e.g. 'mean', 'sum', 'max', 'min', or a callable.
    
    Returns
    -------
    pd.DataFrame with a UTC-normalized DatetimeIndex.
    """
    df = df.copy()
    df[timestamp_col] = pd.to_datetime(df[timestamp_col], utc=True)
    df = df.set_index(timestamp_col)
    return df[value_col].resample(interval).agg(agg_func)


# Usage
hourly = aggregate_by_interval(site_df, 'datetime', 'nitrate_con', '1h')
```

```python
import plotly.express as px

def plot_time_series(df, title='', ylabel=''):
    fig = px.line(
        df.reset_index(),
        x=df.index.name,
        y=df.columns.tolist() if isinstance(df, pd.DataFrame) else df.name,
        title=title,
        labels={'value': ylabel}
    )
    fig.show()

# Usage
hourly = aggregate_by_interval(site_df, 'datetime', 'nitrate_con', '1h')
daily = aggregate_by_interval(site_df, 'datetime', ['nitrate_con', 'turbi_bes'], '1D')
plot_time_series(daily['nitrate_con'], title='Daily Nitrate', ylabel='Nitrate (mg/L)')
plot_time_series(daily['turbi_bes'], title='Daily Nitrate', ylabel='Nitrate (mg/L)')

daily.info()
```

## Accessing some USGS data

```python
import requests
import pandas as pd
from io import StringIO

# Get nitrate data for Iowa USGS sites
url = "https://waterservices.usgs.gov/nwis/iv/"
params = {
    "format": "rdb",
    "stateCd": "ia",
    "parameterCd": "99133",  # nitrate; 00300=DO, 00400=pH, 00010=temp
    "siteStatus": "active"
}
r = requests.get(url, params=params)
# Skip comment lines starting with #
lines = [l for l in r.text.splitlines() if not l.startswith("#") and not l.startswith("5s")]
df = pd.read_csv(StringIO("\n".join(lines)), sep="\t", skiprows=[1])
```
